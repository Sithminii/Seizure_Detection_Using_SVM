```python
import os
import numpy as np
from PyEMD import EMD
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
def compute_id(index):
    string = str(index).zfill(5)
    id_str = 'ID'+string
    return id_str

def sortbyindex(file):
    characters = list(file)
    idx = ''
    digitlist = [item for item in characters if item.isdigit()]
    for digit in digitlist:
        idx += digit
    return int(idx)
```

```python
int_ict_path = '/kaggle/input/data-segmentation-bsp-project/interictal_ictal.npy'
int_ict_label_path = '/kaggle/input/data-segmentation-bsp-project/interictal_ictal_labels.npy'

norm_ict_path = '/kaggle/input/data-segmentation-bsp-project/normal_ictal.npy'
norm_ict_label_path = '/kaggle/input/data-segmentation-bsp-project/normal_ictal_labels.npy'

int_ict_data = np.load(int_ict_path)
int_ict_label = np.load(int_ict_label_path)

norm_ict_data = np.load(norm_ict_path)
norm_ict_label = np.load(norm_ict_label_path)


int_ict_feats = []
norm_ict_feats = []

emd = EMD()

for i in range(len(int_ict_label)):
    feats = []
    
    segment = int_ict_data[i]
    IMFs = emd(segment)
    
    mean_arr = np.mean(np.abs(IMFs[:5,:]), axis=1)
    std_arr = np.std(IMFs[:5,:], axis=1)
    
    Vc = np.power(std_arr,2)/np.power(mean_arr,2)
    feats.append(Vc)
    
    FI = np.sum(np.diff(IMFs[:5, :], axis=1), axis=1)/(1024)
    feats.append(FI)
    
    int_ict_feats.append(feats)


    
    
for i in range(len(norm_ict_label)):
    feats = []
    segment = norm_ict_data[i]
    IMFs = emd(segment)
    
    mean_arr = np.mean(np.abs(IMFs[:5,:]), axis=1)
    std_arr = np.std(IMFs[:5,:], axis=1)
    
    Vc = np.power(std_arr,2)/np.power(mean_arr,2)
    feats.append(Vc)
    
    FI = np.sum(np.diff(IMFs[:5, :], axis=1), axis=1)/(1024)
    feats.append(FI)
    
    norm_ict_feats.append(feats)

    
    
int_ict_feats = np.array(int_ict_feats)
norm_ict_feats = np.array(norm_ict_feats)
```

```python
folder = '/kaggle/working/'
```

```python
np.save(os.path.join(folder,'int_ict_feats.npy'), int_ict_feats)
np.save(os.path.join(folder,'norm_ict_feats.npy'), norm_ict_feats)

np.save(os.path.join(folder,'int_ict_label.npy'), int_ict_label)
np.save(os.path.join(folder,'norm_ict_label.npy'), norm_ict_label)
```
